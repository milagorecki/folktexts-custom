Update file paths in json results file to point to correct file

```python
import glob
folder = '0-bullet-is'
for json_file in glob.glob((ROOTDIR/folder/'model-*/*_task-*/*_bench*/*.json').as_posix()):
    print(Path(json_file).parent.parent.name) #model and task
    file = load_json(json_file)
    # check if paths are correct
    pred_path = Path(file['predictions_path'])
    if not pred_path.is_file():
        new_path =  Path(json_file).parent/pred_path.name
        # only use path from result folder onwards
        new_path = Path(new_path.as_posix()[new_path.as_posix().index('results/'):])
        if new_path.is_file():
            file['predictions_path'] = new_path.as_posix()
            print('updated predictions_path: ', new_path.as_posix())
    # else: 
    #     new_path = Path(pred_path.as_posix()[pred_path.as_posix().index('results/'):])
    #     print(pred_path == new_path)
        # print(f"old pred path: {pred_path}")
        # print(f"potentially new: {new_path}")

    # leave results dir and results root dir as they are (pointing to original dir)
    # results_dir = Path(file['results_dir'])
    # results_root_dir = file['results_root_dir']

    img_paths = file["plots"] #make path relative
    for (key, path) in img_paths.items():
        if not Path(path).is_file():
            new_path =  Path(json_file).parent/Path(Path(path).parent.name)/Path(path).name
            # only use path from result folder onwards
            new_path = Path(new_path.as_posix()[new_path.as_posix().index('results/'):])
            if new_path.is_file():
                file['plots'][key] = new_path.as_posix()
                print('updated img paths: ', new_path.as_posix())
    # save_json(file, json_file)
```
